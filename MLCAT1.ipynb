```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy
from sklearn.preprocessing import StandardScaler
```

df = pd.read_csv(r'C:\Users\Admin\Desktop\ETHICS DATASETS.csv')
print(df.head())

```python
df.info()
```

```python
print("Missing values per column:\n", df.isnull().sum())

numeric_columns = df.select_dtypes(include=['float64', 'int64']).columns

df[numeric_columns] = df[numeric_columns].fillna(df[numeric_columns].mean())

print("Missing values per column after filling:\n", df.isnull().sum())
```

```python
df.describe
```

```python
BOXPLOT
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(data=df[['Inventory Turnover Ratio', 'Lead Time (days)', 'Revenue Growth Rate out of (15)']])
plt.xticks(rotation=45)
plt.title('Boxplot of the Data')
plt.show()
```

CORRELATION HEATMAP

```python
df_numeric = df.select_dtypes(include=[float, int])
plt.figure(figsize=(12, 8))
sns.heatmap(df_numeric.corr(), annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Pie chart for SCM Practices distribution
scm_practices = df['SCM Practices'].value_counts()
plt.figure(figsize=(8, 8))
scm_practices.plot.pie(autopct='%1.1f%%', startangle=90, cmap='Set3')
plt.title('Distribution of SCM Practices')
plt.ylabel('')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.violinplot(data=df, x='SCM Practices', y='Customer Satisfaction (%)')
plt.xticks(rotation=45)
plt.title('Customer Satisfaction by SCM Practices')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='SCM Practices', y='Revenue Growth Rate out of (15)', hue='SCM Practices', palette='Set1')
plt.title('Revenue Growth Rate by SCM Practices')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Inventory Turnover Ratio', y='Customer Satisfaction (%)', hue='SCM Practices', palette='Set2')
plt.title('Inventory Turnover vs Customer Satisfaction')
plt.show()
```
